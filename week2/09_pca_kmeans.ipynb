```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
iris = pd.read_csv('https://raw.githubusercontent.com/devdio/datasets/main/iris.csv')
iris
```

```python
iris['Species'].value_counts()
```

```python
iris.isna().sum()  
```

```python
fig, ax = plt.subplots(2,2, figsize=(10, 10))
sns.boxplot(x='Species', y='SepalLengthCm', data=iris, ax=ax[0,0])
sns.boxplot(x='Species', y='SepalWidthCm', data=iris, ax=ax[0,1])
sns.boxplot(x='Species', y='PetalLengthCm', data=iris, ax=ax[1,0])
sns.boxplot(x='Species', y='PetalWidthCm', data=iris, ax=ax[1,1])
```

```python
X = iris.drop(columns=['Species', 'Id'])
y = iris['Species']

print(X.shape)
print(y.shape)
```

```python
from sklearn.preprocessing import  StandardScaler

scaler = StandardScaler()
scaler.fit(X)
X_scale = scaler.transform(X)

X_scale
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scale)

X_pca
```

```python
dumy = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
dumy
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_encode = le.fit_transform(y)
y_encode
 
```

```python
sns.scatterplot(x='PC1', y='PC2', data=dumy, hue=y_encode, hue_order=[0,1,2], palette='Set1')
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3)
kmeans.fit(X_pca)

sns.scatterplot(x='PC1', y='PC2', data=dumy, hue=kmeans.labels_, palette='Set1')
```

```python
inertia = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X_pca)
    inertia.append(kmeans.inertia_)
    sns.scatterplot(x='PC1', y='PC2', data=dumy, hue=kmeans.labels_, palette='Set1')
    plt.title(f'k={k}')
    plt.show()
    print(k, kmeans.inertia_)

```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(2, 10), inertia, 'o')
plt.plot(range(2, 10), inertia, '--')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```
