```python
import pandas as pd

df = pd.read_csv('data/서울시 공중화장실 위치정보.csv', encoding='cp949')
df
```

```python
# 공중화장실 위도, 경도 정보를 folium의 CircleMarker로 표시

seoul_map = folium.Map(location=[37.55, 126.98], zoom_start=12)

for lat, lng in zip(df['위도'], df['경도']):
    folium.CircleMarker([lat, lng],
                       radius=1,
                       color='blue',
                       fill=True,
                       fill_color='blue',
                       fill_opacity=0.9
                       ).add_to(seoul_map)

seoul_map
```
