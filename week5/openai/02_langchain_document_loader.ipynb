```python
# csv 파일 분석 시키기

from langchain.agents import AgentType
from langchain.llms import AzureOpenAI
from langchain_experimental.agents.agent_toolkits.pandas.base import create_pandas_dataframe_agent
import pandas as pd

llm = AzureOpenAI(deployment_name='dev-gpt-35-turbo-instruct')
df = pd.read_csv('data/titanic3.csv')
agent = create_pandas_dataframe_agent(
    llm=llm,
    df=df,
    agent_type=AgentType.ZERO_SHOT_REACT_DESCRIPTION, # Zero-Shot
    verbose=True # 중간 과정 보여줄지
)
```

```python
query = '데이터를 정리해서 알려줘'
agent.run(query)
```

```python
query = '전체 승객 중 3등급에 탑승한 남성의 생존률을 구해줘'
agent.run(query)
```
